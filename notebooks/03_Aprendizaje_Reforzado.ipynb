## Importar dependencias

````python
import gym 
from stable_baselines3 import PPO
from stable_baselines3.common.vec_env import VecFrameStack
from stable_baselines3.common.evaluation import evaluate_policy
import os
````

````python
environment_name = "CarRacing-v0"
````

Se debe instanciar un ambiente: Esto es la forma de evaluar el desempeño del carrito dependiendo de donde está ahora

````python
env = gym.make(environment_name)
````

Primera ejecución sin entrenar, el carrito se sale

````python
episodes = 5
for episode in range(1, episodes+1):
    state = env.reset()
    done = False
    score = 0 
    
    while not done:
        env.render()
        action = env.action_space.sample()
        n_state, reward, done, info = env.step(action)
        score+=reward
    print('Episode:{} Score:{}'.format(episode, score))
env.close()
````

## Instanciación Modelo a emplear
PPO: ¿Cómo podemos dar el mayor paso de mejora posible en una política utilizando los datos que tenemos actualmente, sin llegar tan lejos que accidentalmente provoquemos un colapso del rendimiento? basicamente emplea algoritmos de primer orden para penalizar durante los entrenamientos los pasos o deltas que se hacen en los ajustes, la idea es que el paso siguiente aumente o disminuya en su escala de cambio en base a que se detecte recompensa mayor o menor.

````python
model = PPO("CnnPolicy", env, verbose=1, tensorboard_log=log_path)
````

## Entrenamiento de modelo

# OJO
Este entrenamiento toma 2 horas

````python

# total_timesteps puede ser ajustado a valores mas pequeños, con 200000 iteraciones nos tomó 2 horas en una maquina con i7, 8 nucleos
model.learn(total_timesteps=200)
````

## Evaluar el modelo despues de 200000 iteraciones

Esto si se entrenó con 200000 iteraciones, con menos va a notar que el automovil se sale mas frecuentemente o se queda girando

````python
evaluate_policy(model, env, n_eval_episodes=5, render=True)
````
